```python
import os
import load
import pandas as pd
import numpy as np
import seaborn as sns
import datetime
```

```python
n_list = [5, 10, 50, 100, 500, 1000, 5000, 10000, 20000, 40000, 80000, 160000]

n_nodes_list = list()
n_edges_list = list()
loading_time = list()

id_trajectories = load.load_trajectory("/home/ubuntu/Data/all_train_irl.csv")

import random

for i in n_list:
    trajectories = random.sample(list(id_trajectories.values()), i)
    starttime = datetime.datetime.now()
    m_graph = load.load_graph_traj(trajectories)
    endtime = datetime.datetime.now()
    n_nodes_list.append(m_graph.get_node_number())
    n_edges_list.append(m_graph.get_edge_number())
    loading_time.append(endtime-starttime)
```

```python
df = pd.DataFrame(np.array([n_list,n_nodes_list,n_edges_list,loading_time]).T, columns=['n_experts','n_nodes','n_edges','loading_time'])
df
# df.plot(x='n_experts', y='n_nodes')

# df.plot(x='n_experts', y='n_edges')

# df.plot(x='n_nodes', y='n_edges')
```
